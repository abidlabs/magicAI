```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
```

```python
# plt.style.use('grayscale')
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']

# data to plot
n_groups = 6
means_frank = (90, 55, 40, 65, 6, 45)
means_guido = (30, 10, 24, 53, 10, 43)

# create plot
fig, ax = plt.subplots(figsize=[8, 3], facecolor='white')

index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_frank, bar_width,
alpha=opacity,
label='Training')

rects2 = plt.bar(index + bar_width, means_guido, bar_width,
alpha=opacity,
label='Deployment')

plt.xlabel('Feature')
plt.ylabel('Proportion (%)')
plt.title('Most differing features')
plt.xticks(index + 0.5*bar_width, ('Pclass=1', 'Sex=Male', 'Pclass=2', 'mean(Ticket)', 'missing(Cabin)', 'mean(Age)'))
plt.legend()


plt.tight_layout()
plt.show()
```
